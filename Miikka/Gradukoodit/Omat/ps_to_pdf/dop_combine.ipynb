### .ps kuvien yhdistämistä varten tehty scripti

##### Doppler tomografia antoi ulos 3 .ps muotoista kuvatiedostoa, jotka tahdoin yhdistää yhdeksi .pdf tiedostoksi.

Toimiakseen, tämä vaatii ps2pdf14 toimintoa, eli ghostscript tukea. Käsittääkseni Ubuntu tukee tätä suoraan, mutta en ota siihen kantaa.

Windowsilla pitää nähdä vähän vaivaa:
https://www.ghostscript.com/releases/gsdnld.html asenna uusin ghostscript
Se minne se asentuu (C:\Program Files\gs\), pitää tämän lokaation kansiot bin ja lib lisätä Windowsin Enviromental Variables systeemin PATH:iin. Tarkemmin tietoa tästä Githubissa.

Toimiakseen, aseta 3 .ps kuvaasi tiettyyn alakansioon. Esimerkiksi, H-alpha .ps kuvat ovat kansiossa H_alpha, ja heliumin 5875Å kuvat kansiossa He_5875. Scripti yrittää sitten yhdistää ne.

### Kiinnitä huomiota riveihin 48-52. Voit säätää niissä kuvien asettelua.

```python
from PyPDF2 import PdfMerger
from PyPDF2 import PdfReader, PdfFileWriter
from PyPDF2 import PageObject
from PyPDF2 import Transformation
from PyPDF2 import PdfWriter

import glob
import os
```

```python

pdf_folders = glob.glob("*")
#print(pdf_folders)
for i in pdf_folders:
    ps_files = glob.glob(f"{i}\\*.ps")
    if ps_files == []:
            print("Not a folder")
            continue
    pdf_files =[]
    for j in ps_files:
        print(j)
        name = f"{j.split('.')[0]}.pdf"
        print(name)
        os.system(f"ps2pdf14 -dPDFsettings#/prepress {j} {name}")
        pdf_files.append(name)
    merger = PdfMerger()
    re_ord = 0
    for pdf in pdf_files:
        if re_ord == 0:
            merger.append(pdf_files[1])
        elif re_ord == 1:
            merger.append(pdf_files[0])
        elif re_ord == 2:
            merger.append(pdf_files[2])
        re_ord += 1
    merger.write("appended.pdf")
    merger.close()
    for j in pdf_files:
        os.remove(j)
        #print(j)
    #pdf1File = open("appended.pdf",'rb')
    reader = PdfReader("appended.pdf")
    pages = []
    for pageNum in range(len(reader.pages)):
        pageObj = reader.pages[pageNum]
        pages.append(pageObj)
    width = pages[0].mediabox.width*1.65
    height = pages[0].mediabox.height/1.8
    merged_page = PageObject.create_blank_page(None, width, height)
    x = 0
    n=0
    #print(pages)
    for page in pages:
        merged_page_mid = PageObject.create_blank_page(None, width, height)
        merged_page_mid.merge_page(page,False)
        x = float(x) + float(page.mediabox.width)/4.5
        if n == 0:
            merged_page_mid.add_transformation(Transformation().scale(1).translate(-40,-350))
        elif n ==1:
            merged_page_mid.add_transformation(Transformation().scale(1).translate(x-40,-350))
        elif n ==2:
            merged_page_mid.add_transformation(Transformation().scale(1.05).translate(x+85,-353.5))
        merged_page.merge_page(merged_page_mid, False)
        n += 1
    writer = PdfWriter()
    writer.add_page(merged_page)

    with open(f'{i}_append.pdf', 'wb') as f:
        writer.write(f)
    writer.close()
```
